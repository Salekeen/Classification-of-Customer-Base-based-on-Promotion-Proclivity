```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

pd.set_option("display.precision",2)
pd.set_option("display.max_columns",None)

# Graphics in SVG format are more sharp and legible
%config InlineBackend.figure_format = 'svg'

import warnings
warnings.filterwarnings("ignore")

```

```python
data = pd.read_csv('Data/marketing_campaign.csv',sep='\t')
```

### Summerizing *People's* Attributes

```python
data.Year_Birth.plot.hist()
plt.title('Birth Year')
plt.show()
```

```python
data.Education.value_counts().plot.barh()
plt.title("Education Level")
plt.show()
```

```python
data.Income.hist()
plt.title("Income Distribution")
```

```python
data.Income.plot.box(vert=False)
plt.title("Income Boxplot")
plt.show()
```

```python
home = ['Kidhome','Teenhome']
for att in home:
    data[att].value_counts().plot.barh()
    plt.tight_layout()
    plt.title(f'{att}')
    plt.show()
```

```python
data.Recency.hist()
plt.title("Recency")
plt.show()
```

```python
data.Complain.value_counts().plot.bar()
plt.title("Complain")
plt.show()
```

### Summerizing *Products* attributes

```python
product_attr = ['MntWines', 'MntFruits',
                'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
                'MntGoldProds']
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8, 5))

for i,attr in enumerate(product_attr):
    ax = axes[int(i / 3), i % 3]
    sns.histplot(data[attr],ax=ax)
fig.tight_layout()
```

### Summerizing *Promotion's* attributes

```python
data.NumDealsPurchases.hist()
plt.title("Num of Deals Purchases")
plt.show()
```

```python
promotion_attr = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4',
                  'AcceptedCmp5', 'Response']
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8, 5))

for i,attr in enumerate(promotion_attr):
    ax = axes[int(i / 3), i % 3]
    sns.countplot(data[attr],ax=ax)
fig.tight_layout()
```

### Summerizing *Place* attributes

```python
place_attr = ['NumWebPurchases',
              'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 5))

for i,attr in enumerate(place_attr):
    ax = axes[int(i / 2), i % 2]
    sns.histplot(data[attr],ax=ax)
fig.tight_layout()
```
